```python
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
xml_loc = "/Users/laurenbeede/Downloads/usa_00001.xml"
dat_loc = "/Users/laurenbeede/Downloads/usa_00001.dat.gz"
```

```python
def parse_ddi(ddi_path):
    tree = ET.parse(ddi_path)
    root = tree.getroot()
    ns = {"ddi": "ddi:codebook:2_5"}

    layout = []

    for var in root.findall(".//ddi:var", ns):
        name = var.get("name")

        loc = var.find("ddi:location", ns)
        if loc is None:
            continue

        start = int(loc.get("StartPos"))
        width = int(loc.get("width"))

        layout.append({
            "name": name,
            "start": start - 1,   # convert to 0-based indexing
            "end": start - 1 + width
        })

    return pd.DataFrame(layout).sort_values("start")

parse_ddi(xml_loc)
```

```python
def read_ipums_dat(dat_gz_path, layout_df):
    colspecs = list(zip(layout_df["start"], layout_df["end"]))
    names = layout_df["name"].tolist()

    df = pd.read_fwf(
        dat_gz_path,
        colspecs=colspecs,
        names=names,
        compression="gzip",
        dtype=str   # keep everything as string initially
    )
    return df

read_ipums_dat(dat_loc, parse_ddi(xml_loc))
```
